#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Подготовка

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, TimeSeriesSplit, GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.dummy import DummyRegressor
from lightgbm import LGBMModel, LGBMRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

import warnings
warnings.filterwarnings("ignore")

PATH = './'
```

```python
df = pd.read_csv(PATH+'taxi.csv',
                 parse_dates=['datetime'],
                 index_col=['datetime'])
df = df.resample('1H').sum()
```

```python
display(df.head())
display(df.describe())
display(df.info())
display(df.isna().sum())
display(df.index.is_monotonic_increasing)
```

## Анализ

```python
decomposed = seasonal_decompose(df[['num_orders']])
trend, seasonal, resid = decomposed.trend, decomposed.seasonal, decomposed.resid
```

```python
slice_ = 24*14
trend[-slice_:].plot(kind='line')
seasonal[-slice_:].plot(kind='line')
resid[-slice_:].plot(kind='line',
                     figsize=(15, 7),
                     grid=True)
plt.title('seasonal_decompose 1H resampled')
plt.ylabel('decompositions')
plt.legend(['trend', 'seasonal', 'resid'])
plt.show()
```

На графике мы может отчетливо заметить, что была успешно найдена и распознана сезонная компонента, в частности суточная сезонность. Но мы так же можем заметить, что на графике тренда присутствует по крайне мере еще одна нераспознанная сезоннасть - недельная. Попробуем установить ее двумя методами:
 * декомпозировать отдельно уже имеющийся тренд, очищенный от сезонной (внутрисуточной) компоненты;
 * декомпозировать изначальный набор данных, но отсемплированный по более значительному временному промежутку (например, дням).

```python
decomposed = seasonal_decompose(trend.dropna())
trend, seasonal, resid = decomposed.trend, decomposed.seasonal, decomposed.resid
```

```python
slice_ = 24*14
trend[-slice_:].plot(kind='line')
seasonal[-slice_:].plot(kind='line')
resid[-slice_:].plot(kind='line',
                     figsize=(15, 7),
                     grid=True)
plt.title('trend\'s seasonal_decompose 1H resampled')
plt.ylabel('decompositions')
plt.legend(['trend', 'seasonal', 'resid'])
plt.show()
```

Первый способ, очевидно, не принес желанного результата. Повторная декомпозиция лишь повторила предыдущий результат.

```python
decomposed = seasonal_decompose(df[['num_orders']].resample('1D').sum())
trend, seasonal, resid = decomposed.trend, decomposed.seasonal, decomposed.resid
```

```python
slice_ = 7*6*2
trend[-slice_:].plot(kind='line')
seasonal[-slice_:].plot(kind='line')
resid[-slice_:].plot(kind='line',
                     figsize=(15, 7),
                     grid=True)
plt.title('seasonal_decompose 1D resampled')
plt.ylabel('decompositions')
plt.legend(['trend', 'seasonal', 'resid'])
plt.show()
```

```python
slice_ = 7*8
seasonal[-slice_:].plot(kind='line')
resid[-slice_:].plot(kind='line',
                     figsize=(15, 7),
                     grid=True)
plt.title('seasonal_decompose 1D resampled')
plt.ylabel('decompositions')
plt.legend(['seasonal', 'resid'])
plt.show()
```

Мы, действительно, обнаружили искомое - цикличность внутри недели, отраженную локальными пиками вокруг понедельников и пятниц. Однако вклад "дня недели" не столь высок, как вышеотраженный вклад "часа". Данный анализ показывает нам на чем стоит сфокусировать большее внимание при моделировании признаков, в частности 'часы' и 'дни недели', а также соответствующие скользящие. Также подойдет что-нибудь, чтобы отразить месте на тренде - например месяц или номер недели в году.

```python
def make_features(df, max_lag, rolling_means=[]):
    df = df.copy()
    df['month'] = df.index.month
    df['week'] = df.index.week
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        df['lag_{}'.format(lag)] = df['num_orders'].shift(lag)
    
    for rolling_mean in rolling_means:
        df[f'rolling_mean_{rolling_mean}'] = df['num_orders'].shift().rolling(rolling_mean).mean()
        
    df['day_shift'] = df['num_orders'].shift(24)
    
    return df
```

```python
df = make_features(df, 8, [4, 8])
```

## Обучение

```python
X = df.drop('num_orders', axis=1)
y = df['num_orders']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1,
                                                    random_state=42, shuffle=False)
```

```python
X_train.dropna(inplace=True)
y_train = y_train[X_train.index]
```

```python
model_types = {
               'adaboost': AdaBoostRegressor(base_estimator=DecisionTreeRegressor(max_depth=1, random_state=42),
                                              random_state=42),
               'rndforest': RandomForestRegressor(random_state=42, n_jobs=-1),
               'lgb': LGBMRegressor(random_state=42, n_jobs=-1),
               'naive': DummyRegressor(strategy='mean')
              }

models_params = {
                 'adaboost': {'n_estimators': [50, 75],
                              'learning_rate': [1.0, 0.01]},
                 'rndforest': {'n_estimators': [50, 75],
                           'max_depth': np.arange(10, 15+1, 5)},
                 'lgb': {'n_estimators': [50, 75],
                         'max_depth': np.arange(10, 15+1, 5),
                         'learning_rate' : [0.1, 0.01]},
                 'naive': {}
                }
```

```python
%%time
def grid_search(model_types, models_params):
    best_models = {}
    cv_results = {}

    for model_type in model_types:
        model = model_types[model_type]
        grid = GridSearchCV(model, models_params[model_type], cv=TimeSeriesSplit(n_splits=5), scoring='neg_root_mean_squared_error',
                            n_jobs=5, refit=True, return_train_score=False)

        grid.fit(X_train, y_train)
        best_models[model_type] = grid.best_estimator_
        cv_results[model_type] = grid.cv_results_
        print('Лучшая модель в группе {} имеет значение RMSE - {:.2f} и параметры {}'\
              .format(model_type,\
                      -grid.best_score_,\
                      grid.best_params_))
        print()
    return best_models, cv_results

best_models, cv_results = grid_search(model_types, models_params)
```

```python
for model_type in cv_results:
    print(model_type)
    display(pd.DataFrame(cv_results[model_type])\
            [['mean_fit_time', 'mean_score_time', 'mean_test_score', 'params', 'rank_test_score']]\
            .sort_values('rank_test_score')
           )
```

Выберем lightgbm как результирующую модель, дальнейшее тестирование проведем, именно, по ней.

## Тестирование

```python
for best_model in best_models:
    if (best_model=='lgb'):
        print(f'{best_model} rsme - {mean_squared_error(y_test, best_models[best_model].predict(X_test), squared=False):.2f}')
        print(f'{best_model} r2 - {r2_score(y_test, best_models[best_model].predict(X_test)):.2f}')
        print(f'{best_model} mae - {mean_absolute_error(y_test, best_models[best_model].predict(X_test)):.2f}')
        print()
```

```python
feature_imp = pd.Series(best_models['lgb'].feature_importances_, index=X_train.columns)
plt.figure(figsize=(15, 7))
sns.barplot(x=feature_imp, y=feature_imp.index)
plt.title('feature importances')
plt.xlabel('importance')
plt.ylabel('features')
plt.show()
```

```python
plt.figure(figsize=(15, 7))
y_train.plot()
y_test.plot()
plt.title('true y')
plt.legend(['y_train', 'y_test'])
plt.show()

plt.figure(figsize=(15, 7))
y_train.plot()
preds = pd.Series(best_models['lgb'].predict(X_test), index=y_test.index)
preds.plot()
preds.rolling(24).mean().plot()
plt.title('predictions with trendline')
plt.legend(['y_train', 'y_preds', 'y_trend'])
plt.show()
```

```python
plt.figure(figsize=(15, 7))
y_train[-100:].plot()
preds.plot()
y_test.plot()
plt.title('predictions with trendline (closer)')
plt.legend(['y_train', 'y_preds', 'y_test'])
plt.show()
```

Модели(light_gbm) удалось достичь заявленного уровня метрики rsme на тестовой выборке. Как мы можем заметить, "сезонные" переменные hour и day_shift (значение того же самого часа за предыдущий день) дейстительно отличаются большей значимостью. Вопреки ожиданиям month, очевидно, не смогла раскрыть общий тренд, тоже самое можно сказать и о значимости переменной dayofweek.
