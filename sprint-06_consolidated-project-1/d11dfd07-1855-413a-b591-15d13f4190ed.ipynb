```python
df = df[~df['puNumber'].isna()].reset_index(drop=True) # полагаю, на такие жертвы мы готовы пойти
```

```python
for column_value, column_slice in df.groupby(['title', 'director']):
    if len(column_slice) >= 3: # для демонстрации оставил самый выдающийся результат 
        display(column_value, column_slice)
```

**При первоначальном просмотре датафрейма было обнаружено, что один и тот же фильм может появляться в датафрейме по несколько раз, в том числе:**
 * в сразными puNumber (то, что напрашивалось на параметр с уникальным идентификатором) и, как следствие, с разными значениями параметра box_office;
 * с разними датами начала старта проката (в большинстве случаев разница не привышвет считаные дни, но иногда это - годы);
 
**Поэтому для эффективного удаления дубликатов на следующем шагу, а также сохранения всей доступной информации:**
* во всех экземлярах фильма дата проката будет заменена на самую раннюю;
* во всех экземлярах фильма кассовые сборы будут заменены на сумму кассовых сборов по всем "прокатам";
* puNumber оставим без изменений - просто при зачистке дубликатов не будем учитывать этот параметр.

```python
for column_value, column_slice in df.groupby(['title', 'director']):
    if len(column_slice) >= 2:
        df.loc[(df['title'] == column_value[0]) &
               (df['director'] == column_value[1]), 'box_office'] = \
        df.loc[(df['title'] == column_value[0]) &
               (df['director'] == column_value[1]), 'box_office'].sum(min_count=1) # чтобы NaN принудительно не заменялись на 0
        
        df.loc[(df['title'] == column_value[0]) &
               (df['director'] == column_value[1]), 'show_start_date'] = \
        df.loc[(df['title'] == column_value[0]) &
               (df['director'] == column_value[1]), 'show_start_date'].min()
        
        df.loc[(df['title'] == column_value[0]) &
               (df['director'] == column_value[1]), 'refundable_support'] = \
        df.loc[(df['title'] == column_value[0]) &
               (df['director'] == column_value[1]), 'refundable_support'].sum(min_count=1)
        
        df.loc[(df['title'] == column_value[0]) &
               (df['director'] == column_value[1]), 'nonrefundable_support'] = \
        df.loc[(df['title'] == column_value[0]) &
               (df['director'] == column_value[1]), 'nonrefundable_support'].sum(min_count=1)
        
        df.loc[(df['title'] == column_value[0]) &
               (df['director'] == column_value[1]), 'budget'] = \
        df.loc[(df['title'] == column_value[0]) &
               (df['director'] == column_value[1]), 'budget'].sum(min_count=1)
```

На этом этапе нет конкретных инструкций и заданий — поищите интересные закономерности в данных. Посмотрите, сколько выделяют средств на поддержку кино. Проверьте, хорошо ли окупаются такие фильмы, какой у них рейтинг.

**данная часть будет разделена на два раздела, так как некоторая информация доступна только по фильмам с государственной поддержкой и не может быть сравнена с остальной частью выборки. В первом разделе мы рассмотрим анализ "в себе". Ну и собственно во второй части постараемся сравнить фильмы с гоподдержкой с оставшейся частью выборки по имеющимся данным.**

####  Внутренний анализ

```python
df = df.query('show_start_year >= 2015') # отбросим все сомнительные величины
# булевая переменная для обозначения наличия господдержки
df['supported'] = ((~df['refundable_support'].isna() & df['refundable_support'] != 0) | \
                   (~df['nonrefundable_support'].isna() & df['nonrefundable_support'] != 0)) & \
                   (~df['budget'].isna() & df['budget'] != 0)

_ = df \
    .query('supported == True') \
    .groupby('show_start_year')[['budget', 'box_office']].sum() \
    .assign(ratio=lambda x: x['box_office']/x['budget'])

display(_)
_[['budget', 'box_office']].plot(kind='bar', grid=True, figsize=(15, 7),
                                 title='бюджет и сборы фильмов в разбивке по годам')
plt.show()
_['ratio'].plot(kind='bar', grid=True, figsize=(15, 7),
                title='средний коэффициент доходности фильмов в разбивке по годам')
plt.show()

_ = df \
    .query('supported == True') \
    .groupby(['genres_first'])[['budget', 'box_office']].sum() \
    .sort_values(by='budget', ascending=False)[:15] \
    .assign(ratio=lambda x: x['box_office']/x['budget'])
display(_)
_[['budget', 'box_office']].plot(kind='bar', grid=True, figsize=(15, 7),
                                 title='бюджет и сборы фильмов в разбивке по жанрам')
plt.show()
_['ratio'].plot(kind='bar', grid=True, figsize=(15, 7),
                title='средний коэффициент доходности фильмов в разбивке по жанрам')
plt.show()
```

**Как видно из представленных графиков - 2017 и 2018 годы были наиболее успешными с точки зрения окупаемости субсидируемых фильмов и сумели преодолеть значение коэффициента окупаемости в 100%.

**Если же мы посмотрим на разбивку доходности фильмов по жанрам, то обнаружим, что прибыльностью отличились только следующие жанры - "комедия" (окупаемость выше 200%), "мультфильмы" (окупаемостть около 95%), "ужасы" (купаемость около 175%) и "спорт" (окупаемость выше 400%). Так же нельзя не отметить весьма высокую абсолютную доходность по жанру "драма", который в целом не сумел достичь общей окупаемости в 100%, но, возможно, более детальная разбивка по жанрам покажет что-то интересное.

```python
_ = df \
    .query('supported == True') \
    .groupby(['genres_first', 'show_start_year'])[['budget', 'box_office']].sum() \
    .assign(ratio=lambda x: x['box_office']/x['budget']) \
    .sort_values(by='budget', ascending=False)[:15] \


display(_)
_[['budget', 'box_office']].plot(kind='bar', grid=True, figsize=(15, 7),
                                 title='бюджет и сборы фильмов в разбивке по годам и жанрам - отранжировано по бюджету')
plt.show()
_['ratio'].plot(kind='bar', grid=True, figsize=(15, 7),
                title='средний коэффициент доходности фильмов в разбивке по годам и жанрам - отранжировано по бюджету')
plt.show()

_ = df \
    .query('supported == True') \
    .groupby(['genres_first', 'show_start_year'])[['budget', 'box_office']].sum() \
    .assign(ratio=lambda x: x['box_office']/x['budget']) \
    .sort_values(by='box_office', ascending=False)[:15] \


display(_)
_[['budget', 'box_office']].plot(kind='bar', grid=True, figsize=(15, 7),
                                 title='бюджет и сборы фильмов в разбивке по годам и жанрам - отранжировано по сборам')
plt.show()
_['ratio'].plot(kind='bar', grid=True, figsize=(15, 7),
                title='средний коэффициент доходности фильмов в разбивке по годам и жанрам - отранжировано по сборам')
plt.show()

_ = df \
    .query('supported == True') \
    .groupby(['genres_first', 'show_start_year'])[['budget', 'box_office']].sum() \
    .assign(ratio=lambda x: x['box_office']/x['budget']) \
    .sort_values(by='ratio', ascending=False)[:15] \


display(_)
_[['budget', 'box_office']].plot(kind='bar', grid=True, figsize=(15, 7),
                                 title='бюджет и сборы фильмов в разбивке по годам и жанрам - отранжировано по коэффициенту доходности')
plt.show()
_['ratio'].plot(kind='bar', grid=True, figsize=(15, 7),
                title='средний коэффициент доходности фильмов в разбивке по годам и жанрам - отранжировано по коэффициенту доходности')
plt.show()
```
